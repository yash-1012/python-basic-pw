#**Q3 Compare and contrast mutable and immutable objects in Python with examples.**

## Answer:

Mutable vs. Immutable Objects in Python

<br>**Mutable Objects:**

Objects whose values can be changed after creation.

Examples: lists, dictionaries, sets.

Example:

`my_list = [1, 2, 3]`

`my_list.append(4)`

`print(my_list)  # Output: [1, 2, 3, 4]`


<br>**Immutable Objects:**

Objects whose values cannot be changed once created.

Examples: numbers (int, float, complex), strings, tuples.

Example:

`my_string = "hello"`

`new_string = my_string + " world"`

`print(my_string)`

`print(new_string)`

#Key Differences:

**Changeability:** Mutable objects can be modified, while immutable objects cannot.

**Memory Allocation:** Mutable objects modify the existing object, while immutable objects create a new object.

**Efficiency:** Immutable objects are generally more efficient due to memory optimization techniques.

Use Cases:

**Mutable Objects:** Ideal for dynamic data structures that require frequent updates, such as lists and dictionaries. However, be cautious as unintended modifications can lead to bugs.

**Immutable Objects:** Suitable for static data structures where data integrity is crucial. They are often more efficient for frequent operations.


#**Q4 Discuss the different types of operators in Python and provide examples of how they are used.**

#Answer:

In Python, operators are symbols or keywords that perform specific operations on variables or values. Hereâ€™s a look at the main types of operators, along with examples to illustrate their usage.

**1. Arithmetic Operators**
Arithmetic operators are used for mathematical calculations. For instance:

Addition (+): `5 + 3` results in `8`.

Subtraction (-): `5 - 3` gives `2`.

Multiplication (*): `5 * 3` produces `15`.

Division (/): `5 / 3` results in approximately `1.666....`

Floor Division (//): `5 // 3` gives `1`, the integer quotient.
Modulus (%): `5 % 3` results in `2`, the remainder.
Exponentiation (**): `5 ** 3` calculates to `125`.

Example:

`a = 10`

`b = 3`

`print(a + b)   # Output: 13`

`print(a ** b)  # Output: 1000`

**2. Comparison Operators**
Comparison operators are used to compare values, returning either True or False. Some examples include:

Equal to (==): `5 == 3` results in `False`.

Not equal to (!=): `5 != 3` returns `True`.

Greater than (>): `5 > 3` is `True`.

Less than (<): `5 < 3` is `False`.

Greater than or equal to (>=): `5 >= 3` is `True`.

Less than or equal to (<=): `5 <= 3` returns `False`.

Example:

`x = 7`

`y = 4`

`print(x > y)   # Output: True`

`print(x == y)  # Output: False`

**3. Assignment Operators**
Assignment operators assign values to variables and can combine assignment with operations. Some examples are:

Assign (=): `a = 5` assigns 5 to a.

Add and assign (+=): `a += 3` is equivalent to a = a + 3.

Subtract and assign (-=): `a -= 2` is equivalent to a = a - 2.

Multiply and assign (*=): `a *= 3` is equivalent to a = a * 3.

Divide and assign (/=): `a /= 2` is equivalent to a = a / 2.

Example:

`a = 10`

`a += 5  # Now a is 15`

`a *= 2  # Now a is 30`

`print(a)  # Output: 30`

**4. Logical Operators**
Logical operators are used to combine conditions and return True or False. Examples include:

And (and): `(5 > 3 and 6 > 3)` returns True only if both conditions are true.

Or (or): `(5 < 3 or 6 > 3)` returns True if at least one condition is true.

Not (not): `not(5 > 3)` returns False because it reverses the result.

Example:

`x = 7`

`y = 10`

`print(x > 5 and y < 15)  # Output: True`

`print(not(x == 7))       # Output: False`

**5. Bitwise Operators**
Bitwise operators perform operations at the binary level on integers.
Examples include:

AND (&): `5 & 3` gives 1, performing bitwise AND.

OR (|): `5 | 3` results in 7, performing bitwise OR.

XOR (^): `5 ^ 3` results in 6, performing bitwise XOR.

NOT (~): `~5 results in -6, performing bitwise NOT.

Left shift (<<): `5 << 1` shifts the bits of 5 to the left, giving 10.

Right shift (>>): `5 >> 1` shifts the bits to the right, resulting in 2.

Example:

`a = 5   # Binary: 101`

`b = 3   # Binary: 011`

`print(a & b)  # Output: 1 (Binary: 001)`

`print(a | b)  # Output: 7 (Binary: 111)`

**6. Membership Operators**
Membership operators test if a value is in a sequence, such as a list or string. Examples include:

In (in): `"a" in "apple"` returns True.

Not in (not in): `"b" not in "apple"` returns True.

Example:

`fruits = ["apple", "banana", "cherry"]`

`print("banana" in fruits)  # Output: True`

`print("grape" not in fruits)  # Output: True`

**7. Identity Operators**
Identity operators compare if two objects refer to the same memory location. Examples include:

Is (is): `a is b` is True if both variables point to the same object.

Is not (is not): `a is not b` is True if the variables point to different objects.

Example:
`a = [1, 2, 3]`

`b = a`

`c = [1, 2, 3]`

`print(a is b)   # Output: True (both refer to the same list)`

`print(a is c)   # Output: False (different objects with identical contents)`

#**Q5 Explain the concept of type casting in Python with examples?**

Answer:

Type casting, also known as type conversion, is the process of converting one data type to another. Python provides several built-in functions to perform type casting:

**1. int():** Converts a value to an integer.

`x = float(3.14)`

`y = int(x)  # y will be 3`

**2. float():** Converts a value to a floating-point number.

`x = 5`

`y = float(x)  # y will be 5.0`

**3. str():** Converts a value to a string.

`x = 10`

`y = str(x)  # y will be '10'`

**4. bool():** Converts a value to a Boolean (True or False).

`x = 1`
`y = bool(x)  # y will be True`

<br> **Important Considerations:**

**Implicit Type Conversion:** Python often performs implicit type conversion when it's safe to do so, such as when combining integers and floating-point numbers.

`x = 5`

`y = 3.14`

`z = x + y  # z will be 8.14`

**Explicit Type Conversion:** You can explicitly convert data types using the functions mentioned above. This is often necessary when you want to perform specific operations or when implicit conversion might lead to unexpected results.

**Type Errors:** If you attempt to perform operations on incompatible data types without proper type casting, Python will raise a TypeError.

**Conditional Statements in Python**

Conditional statements allow you to control the flow of your program based on specific conditions. Python uses the if, elif, and else keywords to create conditional blocks.

**1. if Statement:**

Executes a block of code if a condition is True.

`x = 10`

`if x > 5:`

          print("x is greater than 5")

**2. if-else Statement:**

Executes one block of code if a condition is True, and another block if it's False.

`x = 10`

`if x > 15:`

        print("x is greater than 15")

`else:`

        print("x is less than or equal to 15")

**3. if-elif-else Statement:**

Allows you to check multiple conditions sequentially.

`x = 10`

`if x > 15:`

        print("x is greater than 15")

`elif x > 5:`

        print("x is greater than 5 but less than or equal to 15")

`else:`

        print("x is less than or equal to 5")


**Key Points:**

**Indentation:** Python uses indentation to define code blocks within conditional statements.

**Boolean Expressions:** Conditions are evaluated to either True or False.
Logical Operators: and, or, and not can be used to combine multiple conditions.
